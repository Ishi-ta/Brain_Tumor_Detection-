```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread(r'C:\Users\hp\OneDrive\Desktop\Brain tumor minor\test_images\0 (239).jpg',0)
print(img.shape)
height,width= img.shape
plt.imshow(img,cmap='gray')
```

```python
imgflatten= img.reshape(height*width,1)
print(imgflatten.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6)
model.fit(imgflatten)
labels = model.labels_
print(labels.shape)
label2d =labels.reshape(height,width)
```

```python
plt.imshow(label2d)
plt.colorbar()
plt.show()
```
